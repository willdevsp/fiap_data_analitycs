```python
import pandas as pd
import matplotlib.pyplot as plt
dados = pd.read_csv('mall.csv')
```

```python
dados.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dados['Gender'] = le.fit_transform(dados['Gender'])

dados.head()


```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6, random_state=0)

kmeans.fit(dados[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
kmeans_labels = kmeans.predict(dados[['Annual Income (k$)', 'Spending Score (1-100)']])

centroides = kmeans.cluster_centers_
print(centroides)

kmeans_labels
```

```python
dados.head()
```

```python
k = list(range(1,10))
print(k)
```

```python
sse = []

for i in k:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(dados[['Annual Income (k$)', 'Spending Score (1-100)']])
    sse.append(kmeans.inertia_)


plt.rcParams['figure.figsize'] = (10,5)

plt.plot(k, sse, '-o')
plt.xlabel(r'Número de cluster')
plt.ylabel('Inércia')
plt.show()
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=10, min_samples=8)

dbscan.fit(dados[['Annual Income (k$)','Spending Score (1-100)']])

dbscan_labels = dbscan.labels_
dbscan_labels
```

```python

plt.scatter(dados[['Annual Income (k$)']], dados[['Spending Score (1-100)']], c=dbscan_labels, cmap='rainbow', alpha=0.5)
plt.xlabel('Salario anual')
plt.ylabel('Pontuação de gastos')
plt.scatter(centroides[:, 0], centroides[:, 1], c='Black', marker='X', s=200, alpha=0.5)
#plt.rcParams('figure.figsize') = (10,5)
plt.show()
```

```python
#plot sem os outliers

mascara = dbscan_labels >= 0

plt.scatter(dados[['Annual Income (k$)']][mascara], dados[['Spending Score (1-100)']][mascara], c=dbscan_labels[mascara], cmap='rainbow', alpha=0.5)
plt.xlabel('Salario anual')
plt.ylabel('Pontuação de gastos')
#plt.rcParams('figure.figsize') = (10,5)
plt.show()
```

# Como validar uma clusterização
Temos dois tipos:
- interna: quanto foi bom meu agrupamento
- externa: como parecido estão meus dois algoritmos comparados?

# Avaliando o desempenho dos algoritmos

## Tipo externo:
(a) Usando o **Adjusted Rand Index**
Compara o desempenho quando forem fornecidos datasets com labels geradas de forma aleatória. Quando essas labels estão muito diferente, o valor se aproxima de 0, o que sugere um resultado negativo, ou seja, cluster não próximos

```python
from sklearn.metrics import adjusted_rand_score
```

```python
adjusted_rand_score(kmeans_labels, dbscan_labels)
```

# Tipo interno
(b) avalia a metrica de **silhouette**
Mede o formato do cluster obtido: avalia a distância entre os centros dos clusters, nesse caso, queremos maximizar as disctancias
valores próximo a -1, significa clusters ruins, próximo a 1, cluster bem parecidos

```python
from sklearn.metrics import silhouette_score
```

## KMEANS

```python
silhouette_score(dados[['Annual Income (k$)','Spending Score (1-100)']],kmeans_labels)
```

## DBSCAN

```python
silhouette_score(dados[['Annual Income (k$)','Spending Score (1-100)']],dbscan_labels)
```
