## Colocar uma grade (grid) horizontal e vertical que não seja intrusivo

```python
def plota_gastos_por_ano_com_grid(dados):
    cores = [   "#b2df8a", "#fb9a99", "#fdbf6f", "#cab2d6", "#ffff99", "#1f78b4", "#33a02c", "#e31a1c", "#ff7f00", "#1f78b4", "#33a02c", "#e31a1c", "#ff7f00", "#6a3d9a", "#b15928", "#a6cee3", "#6a3d9a" ]   
    new_colors = {}
    new_colors = dict(zip(dados.index, cores))
    #for coluna, cor in zip(dados.index, cores):
    #    new_colors[coluna]=cor

    
    
    axis = dados.head(8).T.plot(figsize=(20,10), color = new_colors,grid= True)
    
    axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
    axis.set_ylabel("Gastos aprovados em milhões de reais")
    axis.set_xlabel("Ano")
    
    
    plt.show()
    

plota_gastos_por_ano_com_grid(ordenados_por_total.filter(regex="Jan$"))
```
